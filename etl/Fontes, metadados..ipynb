```python
for r in df.iterrows():
    codigo = r[1][1].upper()
    try:
        fontes = r[1][6].split(';')
    except AttributeError:
        continue
    for fonte_name in fontes:
        try:
            fonte = Fonte.objects.get(nome=fonte_name)
        except Fonte.DoesNotExist:
            print u"fonte '{}' não encontrada.".format(fonte_name)
            fonte = Fonte.objects.create(nome=fonte_name)
        for i in Indicador.objects.filter(observacao__icontains=codigo).exclude(fontes=fonte):
            if codigo in [_ for _ in i.observacao.upper().split(',')]:
                i.fontes.add(fonte.pk)
                print i, fonte
```
